```python
text = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. 
On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. The highe daylight temperature of the Moon is 127 C. """
print(text)
dividirTexto=text.split('.')
print(dividirTexto)
palabraC=['Moon']
for recorido in dividirTexto:
    for palabra in palabraC:
        if palabra in  recorido:
            print(recorido)
            break
for recorido in dividirTexto:
    for palabra in palabraC:
        if palabra in recorido:
            print(recorido.replace(' C','Celsius'))
            break
```
