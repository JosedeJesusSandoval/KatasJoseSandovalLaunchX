```python
def mision(tiempoA,tiempoV,destino,tanque2,tanque1):
    return f"""
    Destino: {destino}
    Tiempo en llegar: {tiempoA+tiempoV} minutos
    Tanque de gasolina: {tanque1+tanque2} galones"""
print(mision(14,51,"Luna",200000,300000))
def mision_Actualizacion(destino,*minutos,**DepositoC):
    return f"""
    Destino: {destino}
    Tiempo en llegar: {sum(minutos)} minutos
    Tanque de gasolina: {sum(DepositoC.values())} galones
    """
print(mision_Actualizacion("Moon",10,15,51,depos1=300000,depos2=200000))
def mision_Actua2(destino,*minutos,**DepositoC2):
    principal_report= f"""
    Destino: {destino}
    Tiempo en llegar: {sum(minutos)} minutos
    Tanque de gasolina: {sum(DepositoC2.values())} galones
    """
    for tanque,galones in DepositoC2.items():
        principal_report += f"""{tanque} tanque---> {galones} galones"""
    return principal_report
print(mision_Actua2("Luna", 8, 11, 55, main=300000, external=200000))
```
