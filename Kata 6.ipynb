```python
planets=['Mercurio','Venus','Tierra','Marte','Júpiter','Saturno','Urano','Neptuno']
userlanet=input("Dime el nombre del planeta comenzando con una mayuscula y depues minusculas: ")
usuario=planets.index(userlanet)
planets_after=planets[0:usuario]
print("Los planetas más cercanos al sol son: ",planets_after)
planets_berfore=planets[usuario+1:]
print("Los planetas más alejados del sol son: ",planets_berfore)
```
