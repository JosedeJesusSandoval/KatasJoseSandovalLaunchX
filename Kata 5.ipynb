Ejercicio 1

```python
from math import ceil
tierra=149597870 
jupiter=778547200 
distancia=jupiter-tierra
print("""La distancia entre la tierra y jupiter es de: %s km""" %(distancia))
print("La distancia en millas es: %s millas"%(ceil(distancia*0.621)))
```

Ejercicio 2

```python
from math import ceil
planeta1=input('Dime la distancia del primer planeta en Km: ')
planeta2=input('Dime la distancia del segundo planeta en km: ')
distancia=int(planeta1)-int(planeta2)
print("""La distancia entre la tierra y jupiter es de: %s km""" %(abs(distancia)))
print("La distancia en millas es: %s millas"%(abs(ceil(distancia*0.621))))
```
